```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading the dataset to a pandas DataFrame

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# checking the no. of missing values in each column
credit_card_data.isnull().sum()
```

```python
#Distribution of legit transaction & fraudlent transactions
credit_card_data['Class'].value_counts()
```

```python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Compare the values for both tansactions
credit_card_data.groupby('Class').mean()
```

```python
#Under Sampling
# build a sample dataset containing similar distribution of normal transactions and fraudlent transactions
# Number of Fraudulent Transactions --> 2
legit_sample = legit.sample(n=2)
```

```python
#Concatenating two DataFrames
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python

new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into Features & Targets
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python

print(Y)
```

```python
#Split the data into Training data & Testing Data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2) # remove stratify
```

```python
print(X.shape, X_train.shape, X_test.shape)

```

```python
#Model Training
#Logistic Regression
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation
Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
